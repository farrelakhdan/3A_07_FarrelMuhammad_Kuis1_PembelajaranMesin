```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Raw Data
df = pd.read_csv('insurance.csv')
df.head()
```

```python
# Labeling
label_sex = {
    'male': 1,
    'female': 0
}

label_smoker ={
    'yes' : 1,
    'no' : 0
}

label_region ={
    'northwest' : 0,
    'southwest' : 1,
    'southeast' : 2,
    'northeast' : 3 
}

df['sex'] = df['sex'].map(label_sex)
df['smoker'] = df['smoker'].map(label_smoker)
df['region'] = df['region'].map(label_region)


# Data After Labeling 
df.head()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

#Dataset
X = df[['age', 'bmi', 'region','sex', 'smoker', 'children']]
  
# VIF DF
data_vif = pd.DataFrame()
data_vif["feature"] = X.columns
  
# Calculating VIF DF
data_vif["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(data_vif)
```

```python
# Dropping Column
df = df.drop(df.columns[0], axis=1)
df = df.drop(df.columns[1], axis=1)
df.head()
```

```python
# Data After Processing

new_data = df
new_data.head()
```

```python
# Regressing
X = new_data.iloc[:, :-1].values
y = new_data.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
y = y.reshape(len(y), 1)
# Check Shape
y.shape
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
# Mixing
gabung = np.concatenate((y_test, y_pred))
gabung
```

```python
# Get R2 Score
r2_lr = r2_score(y_test, y_pred)
print(r2_lr)
```
